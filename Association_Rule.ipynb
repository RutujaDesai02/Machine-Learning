```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import association_rules,apriori
import warnings
warnings.filterwarnings('ignore')
```

```python
data.drop_duplicates(subset=['Products'],keep='first',inplace=True)
```

```python
def txt_split(txt):
    return txt.split(',')
```

```python
df = data['Products'].apply(txt_split)
```

```python
df
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
encoded_df = te.fit_transform(df)
```

```python
encoded_df
```

```python
te.columns_
```

```python
df = pd.DataFrame(encoded_df, columns=te.columns_)
df
```

```python
df.replace([True,False],[1,0], inplace=True)
```

```python
df
```

```python
df.shape
```

## Apriori algorithm

```python
scores = apriori(df,min_support=0.1, use_colnames=True)
scores
```

```python
rules = association_rules(scores,metric='lift',min_threshold=0.1)
rules
```

```python
rules.sort_values(by='lift',ascending = False)
```

```python
score1 = apriori(df,min_support=0.2, use_colnames=True)
score1
```

```python
rule1 = association_rules(score1,metric='lift',min_threshold=0.2)
rule1
```

```python
rule1.sort_values(by='lift',ascending = False)
```
